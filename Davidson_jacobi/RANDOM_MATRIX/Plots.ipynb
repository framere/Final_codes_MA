```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set(style="whitegrid")
sns.set_context("notebook")

matrices = [f"A{i}" for i in range(1, 7)]
odd_numbers = 20*(10000)**3
even_numbers = 20*(20000)**3
flops_exact = np.array([odd_numbers, even_numbers, odd_numbers, even_numbers, odd_numbers, even_numbers])
flops_davidson = np.array([19634098930660, 251611935895600, 103243812008660, 989072830408400, 171401308803640, 1375833673055660])

x = np.arange(len(matrices))
width = 0.35

plt.bar(x - width/2, flops_exact, width, label='Exact')
plt.bar(x + width/2, flops_davidson, width, label='Davidson')

plt.yscale('log')
plt.ylabel('FLOPs (log scale)')
plt.xlabel('Matrix')
plt.title('FLOP Comparison: Exact vs Davidson')
plt.xticks(x, matrices)
plt.legend()
plt.tight_layout()
plt.show()
```
