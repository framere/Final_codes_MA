```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

# $H_2$

```python
beta = [16, 25, 32, 36, 40]

# FLOPs_10 = np.array([274706911194, 672060980022, 862867465880, 4380870706860])
FLOPs_50 = np.array([672060980022, 927787337668, 968899750776])
FLOPs_100 = np.array([862867465880, 1125803501460, 1909286762342])
FLOPs_200 = np.array([4380870706860, 1714328837080, 2068221607504])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(beta[:3], FLOPs_50, marker='o', label='l=50')
plt.plot(beta[:3], FLOPs_100, marker='o', label='l=100')
plt.plot(beta[:3], FLOPs_200, marker='o', label='l=200')
plt.yscale('log')
plt.xlabel('Beta')
plt.ylabel('FLOPs (log scale)')
plt.title('FLOPs vs Beta for Different l Values')
xticks = [16, 20, 25, 32, 36, 40]
plt.xticks(xticks)
plt.legend()
plt.grid(True)
plt.show()
```

# Formaldehyde


```python
beta = [16, 20, 25, 32]

FLOPs_10 = np.array([4148365720086, 4372274410042])
FLOPs_50 = np.array([25426240907522, 25786871933884])
FLOPs_100 = np.array([38471656220710, 35723179595297])
FLOPs_200 = np.array([68828434376950, 80876669157091])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(beta[2:], FLOPs_10 / 1e12, marker='o', label='l=10')
plt.plot(beta[2:], FLOPs_50 / 1e12, marker='o', label='l=50')
plt.plot(beta[2:], FLOPs_100 / 1e12, marker='o', label='l=100')
plt.plot(beta[2:], FLOPs_200 / 1e12, marker='o', label='l=200')
plt.yscale('log')
plt.xlabel('Beta')
plt.ylabel('FLOPs (log scale, in trillions)')
plt.title('FLOPs vs Beta for Different l Values (Formaldehyde)')
xticks = [16, 20, 25, 32]
plt.xticks(xticks)
plt.legend()
plt.grid(True)
plt.show()
```
