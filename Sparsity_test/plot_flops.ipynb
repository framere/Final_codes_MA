```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")
sns.set_context("notebook")
colors = sns.color_palette("tab10")
```

```python
N = [100, 500, 1000, 2000]

flops_25965 = [
    8482261419200,
    26104795485760,
    52437276564640,
    125738055377500
]

flops_25966 = [
    8289010152160,
    24449671145400,
    49647729420700,
    109739415009120
]

flops_25967 = [
    5275553867920,
    21329280540000,
    44080967314560,
    106893809407520
]

flops_25968 = [
    5296475360740,
    20238070730880,
    43835653860480,
    85720601980000
]

flops_25969 = [
    5425640964620,
    20073413631840,
    35635220840000,
    85746756220000
]

flops_25971 = [
    9977622638240,
    21536781735040,
    35643610760000,
    85771491820000
]


full_diago_flops = 20 * 30000**3
```

```python
plt.figure(figsize=(7, 5))
plt.axhline(y=full_diago_flops, color=colors[3], linestyle='--', label='Exact Diagonalization')
plt.plot(N, flops_25965, marker='d', label=r'$\eta = 10^{2}$', color=colors[0])
plt.plot(N, flops_25966, marker='o', label=r'$\eta = 10^{4}$', color=colors[1])
plt.plot(N, flops_25967, marker='s', label=r'$\eta = 10^{6}$', color=colors[2])
plt.plot(N, flops_25968, marker='^', label=r'$\eta = 10^{8}$', color=colors[4])
plt.plot(N, flops_25969, marker='x', label=r'$\eta = 10^{10}$', color=colors[5])
plt.plot(N, flops_25971, marker='v', label=r'$\eta = 10^{12}$', color=colors[6])
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Number of computed eigenvalues')
plt.ylabel('FLOPS')
N_occ = 10
xticks = np.array([10, 50, 100, 200]) * N_occ
# Create labels: one line for nev, second line for multiplier
xtick_labels = [f"{mult}$N_{{occ}}$" for nev, mult in zip(xticks, [10, 50, 100, 200])]
plt.xticks(xticks, xtick_labels)
plt.title(r'FLOPS Comparison for test molecule (${N_{occ}} = 10$)')
plt.legend()
plt.grid(True)
sns.despine()
plt.tight_layout()
plt.savefig('FLOPS_comparison_test_molecule.pdf')
plt.show()
```
