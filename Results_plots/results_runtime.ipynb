```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

colors = sns.color_palette()
sns.set(style='whitegrid')
```

# $H_2$

```python
N_occ =1
N = np.array([10, 50, 100, 200])
N_JD = np.array([10, 15, 25, 50, 100, 200])

exact_diago = 320.407197 

# Block Davidson (alpha=8)
time_8_BD = np.array([11.601992, 29.426218, 4280.831207, 19683.393319])

# Block JD (alpha=8)
block_JD_8 = np.array([37.126752, 88.143065, 185.107921, 369.259726, 1088.925034, 4389.471272])

# deflation with JD
time_25_defl_JD = np.array([2034.836621, 13976.914828, 32755.037764, 78011.364205])

# Deflation algorithm (diag preconditioner)
time_25_defl = np.array([14.938182, 31.867831, 75.197833, 142.814585])

# JD mix 
time_25_defl_mix = np.array([36.214662, 115.868661, 196.404753 , 361.804543])
```

```python
plt.figure(figsize=(8, 6))
plt.plot(N, time_8_BD, marker='o', label='8 Block BD', color=colors[0])
plt.plot(N_JD, block_JD_8, marker='o', label='8 Block JD', color=colors[1])
plt.plot(N, time_25_defl, marker='o', label='25 Defl Diag. Prec.', color=colors[2])
plt.plot(N, time_25_defl_JD, marker='o', label='25 Defl JD', color=colors[3])
plt.plot(N, time_25_defl_mix, marker='o', label='25 Defl Mix', color=colors[4])
plt.axhline(y=exact_diago, color='black', linestyle='--', label='Exact Diagonalization', linewidth=2)
plt.yscale('log')   
plt.xscale('log')
plt.xlabel('Matrix Size (N)', fontsize=14)  
plt.ylabel('Time (seconds)', fontsize=14)
xticks = np.array([10, 50, 100, 200]) * N_occ
# Create labels: one line for nev, second line for multiplier
xtick_labels = [f"{mult}×$N_{{occ}}$" for nev, mult in zip(xticks, [10, 50, 100, 200])]
plt.xticks(xticks, xtick_labels)
plt.title('Runtime Comparison', fontsize=16)
plt.legend(fontsize=12)
plt.tight_layout()
sns.despine()
plt.show()
```
