```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set(style="whitegrid")
sns.set_context("notebook")
odd_numbers = 20*(10000)**3
even_numbers = 20*(20000)**3
```

```python
flops = np.array([
    5422167000000,
    40054858360000,
    7713966805440,
    52889116219360,
    7683516127680,
    52695837864640,
    7729249920000,
    53102387740000,
    7724151261120,
    55356026986720,
    9290412520040,
    113060134548680,
    15798387926340,
    319362327297100
])

flops_smaller = flops[::2]
flops_larger = flops[1::2]
```

```python
larger_system_x = np.ones(len(flops_larger)) * even_numbers
smaller_system_x = np.ones(len(flops_smaller)) * odd_numbers

# flops_larger = np.array([1375833673055660, 989072830408400, 251611935895600,55356026986720, 53102387740000, 52889116219360, 52695837864640, 40054858360000])
# flops_smaller = np.array([171401308803640, 319362327297100, 15798387926340, 7729249920000, 7724151261120, 7683516127680, 7713966805440, 5422167000000])

labels = ['1e3', '1e4', '1e5', '1e6', '1e7', '1e9', '1e11']
labels = labels[::-1]
x = np.arange(len(labels))

width = 0.35

plt.bar(x - width/2, smaller_system_x, width, label='Exact')
plt.bar(x + width/2, flops_smaller, width, label='Davidson')
plt.yscale('log')
plt.ylabel('FLOPs (log scale)')
plt.xlabel('Scaling factor')
plt.title('FLOP Comparison for N = 10000')
plt.xticks(x, labels)
plt.legend()
plt.tight_layout()
sns.despine()
plt.show()
```

```python
plt.bar(x - width/2, larger_system_x, width, label='Exact')
plt.bar(x + width/2, flops_larger, width, label='Davidson')
plt.yscale('log')
plt.ylabel('FLOPs (log scale)')
plt.xlabel('Scaling factor')
plt.title('FLOP Comparison for N = 20000')
plt.xticks(x, labels)
plt.legend()
plt.tight_layout()
sns.despine()
plt.show()
```
