```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

colors = sns.color_palette()
sns.set(style='whitegrid')
```

# $CH_2O$

```python
N_JD = np.array([10, 15, 25, 50, 100, 200]) *6
N_block = np.array([10, 15, 17, 19, 21, 23, 25]) *6
N_defl = np.array([10, 50, 100, 200]) *6

time_25_defl = np.array([198.440127, 1520.764473, 5631.008853, 15861.715734])

time_25_JD = np.array([34.978614, 77.622634, 181.049824, 657.490348, 2453.682518, 9550.769901])

time_8_block = np.array([159.120053, 257.375384, 150.762805, 226.299259, 324.986497, 13973.173651, 19171.135959])

time_8_block_JD = np.array([1787.226035, 7655.709516, 7625.291390, 39994.693763, 83014.988620, 81138.472397])

time_25_defl_mix = np.array([226.329849, 18118.276170, 15663.464728, 9176.008875])

time_40_defl_JD = np.array([2420.401205, 8800.863119, 21032.304864, 78011.364205])
```

```python
plt.figure(figsize=(8, 6))

plt.plot(N_JD, time_25_JD, marker='o', label='25 JD', color=colors[0])
plt.plot(N_block, time_8_block, marker='o', label='8 Block', color=colors[1])
plt.plot(N_defl, time_25_defl, marker='o', label='25 Defl', color=colors[2])
plt.plot(N_JD, time_8_block_JD, marker='o', label='8 Block JD', color=colors[3])
plt.plot(N_defl, time_25_defl_mix, marker='o', label='25 Defl Mix', color=colors[4])
plt.plot(N_defl, time_40_defl_JD, marker='o', label='40 Defl JD', color=colors[5])
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Number of Eigenvalues')
plt.ylabel('Runtime (s)')
plt.title('Runtime vs Number of Eigenvalues for Different Methods')
plt.legend()
# plt.grid(True, which="both", ls="--", linewidth=0.5)
plt.tight_layout()
plt.show()
```
